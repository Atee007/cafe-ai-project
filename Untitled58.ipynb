```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from datetime import timedelta
import matplotlib.pyplot as plt

# --- STEP 1: โหลดข้อมูลจากไฟล์ Excel ---
file_path = '/content/Coffee Shop Sales.xlsx'
# หมายเหตุ: ถ้าในไฟล์มีหลาย Sheet อย่าลืมระบุ sheet_name เช่น sheet_name='Transactions'
df = pd.read_excel(file_path)

print("--- ข้อมูลเบื้องต้น ---")
print(df.head())

# --- STEP 2: ทำความสะอาดและเตรียมข้อมูล (Data Pre-processing) ---
# 1. คำนวณยอดขายรวมต่อรายการ
df['total_sales'] = df['transaction_qty'] * df['unit_price']

# 2. แปลงวันที่ให้ถูกต้อง
df['transaction_date'] = pd.to_datetime(df['transaction_date'])

# 3. รวมยอดขายเป็น "รายวัน" (Daily Sales)
# เราจะพยากรณ์ภาพรวมร้าน ถ้าพี่อยากแยกสาขา ต้องกรอง store_location ก่อนครับ
daily_sales = df.groupby('transaction_date')['total_sales'].sum().reset_index()
daily_sales.columns = ['date', 'sales']
daily_sales = daily_sales.sort_values('date')

# --- STEP 3: สร้าง Feature (ตัวแปรที่ AI จะใช้เรียนรู้) ---
def create_features(df):
    df = df.copy()
    df['day_of_week'] = df['date'].dt.dayofweek
    df['month'] = df['date'].dt.month
    df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

    # ยอดขายย้อนหลัง 1 วัน และ 7 วัน (เพื่อให้ AI รู้แนวโน้ม)
    df['sales_lag1'] = df['sales'].shift(1)
    df['sales_lag7'] = df['sales'].shift(7)

    # ค่าเฉลี่ยเคลื่อนที่ 7 วัน
    df['rolling_mean_7'] = df['sales'].shift(1).rolling(window=7).mean()

    return df.dropna()

data_prepared = create_features(daily_sales)

# --- STEP 4: สร้างโมเดลและฝึกสอน (Training) ---
features = ['day_of_week', 'month', 'is_weekend', 'sales_lag1', 'sales_lag7', 'rolling_mean_7']
X = data_prepared[features]
y = data_prepared['sales']

# ใช้ข้อมูลทั้งหมดในการสอนเพื่อความแม่นยำในการพยากรณ์อนาคต
model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(X, y)

print("\n--- สอน AI เรียบร้อยแล้ว ---")

# --- STEP 5: พยากรณ์ยอดขายวันพรุ่งนี้ (Prediction) ---
# ดึงข้อมูลล่าสุดมาเตรียม Feature สำหรับวันถัดไป
last_date = daily_sales['date'].max()
next_date = last_date + timedelta(days=1)

# สร้างข้อมูลจำลองของวันพรุ่งนี้เพื่อส่งให้ AI ทาย
last_sales = daily_sales['sales'].iloc[-1]
last_week_sales = daily_sales['sales'].iloc[-7]
rolling_7 = daily_sales['sales'].tail(7).mean()

next_day_features = pd.DataFrame([{
    'day_of_week': next_date.dayofweek,
    'month': next_date.month,
    'is_weekend': 1 if next_date.dayofweek >= 5 else 0,
    'sales_lag1': last_sales,
    'sales_lag7': last_week_sales,
    'rolling_mean_7': rolling_7
}])

prediction = model.predict(next_day_features)

print(f"\n================================")
print(f"วันที่ล่าสุดในข้อมูล: {last_date.date()}")
print(f"AI คาดการณ์ยอดขายวันที่ {next_date.date()}: {prediction[0]:,.2f} บาท")
print(f"================================")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from datetime import timedelta
import matplotlib.pyplot as plt

# --- เตรียมข้อมูลพื้นฐาน (เหมือนเดิม) ---
df = pd.read_excel('/content/Coffee Shop Sales.xlsx')
df['total_sales'] = df['transaction_qty'] * df['unit_price']
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
daily_sales = df.groupby('transaction_date')['total_sales'].sum().reset_index()
daily_sales.columns = ['date', 'sales']

# --- ฟังก์ชันสร้าง Feature ---
def get_features(date, last_7_days_sales):
    return pd.DataFrame([{
        'day_of_week': date.dayofweek,
        'month': date.month,
        'is_weekend': 1 if date.dayofweek >= 5 else 0,
        'sales_lag1': last_7_days_sales[-1],
        'sales_lag7': last_7_days_sales[0],
        'rolling_mean_7': np.mean(last_7_days_sales)
    }])

# --- ฝึกโมเดล ---
# เตรียมข้อมูลย้อนหลังสำหรับ Train (เหมือนโค้ดก่อนหน้า)
data_train = daily_sales.copy()
data_train['day_of_week'] = data_train['date'].dt.dayofweek
data_train['month'] = data_train['date'].dt.month
data_train['is_weekend'] = data_train['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
data_train['sales_lag1'] = data_train['sales'].shift(1)
data_train['sales_lag7'] = data_train['sales'].shift(7)
data_train['rolling_mean_7'] = data_train['sales'].shift(1).rolling(window=7).mean()
data_train = data_train.dropna()

features_list = ['day_of_week', 'month', 'is_weekend', 'sales_lag1', 'sales_lag7', 'rolling_mean_7']
model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(data_train[features_list], data_train['sales'])

# --- เริ่มพยากรณ์ล่วงหน้า 7 วัน ---
forecast_results = []
current_history = list(daily_sales['sales'].tail(7)) # ดึงยอดขาย 7 วันล่าสุดมาเป็นฐาน
last_date = daily_sales['date'].max()

for i in range(1, 8):
    next_date = last_date + timedelta(days=i)
    features = get_features(next_date, current_history)
    pred_sales = model.predict(features)[0]

    forecast_results.append({'date': next_date, 'predicted_sales': pred_sales})

    # อัปเดตประวัติ: ลบวันแรกสุดออก เพิ่มวันล่าสุดที่พยากรณ์ได้เข้าไป (Recursive)
    current_history.append(pred_sales)
    current_history.pop(0)

forecast_df = pd.DataFrame(forecast_results)

# --- พล็อตกราฟ ---
plt.figure(figsize=(12, 6))
plt.plot(daily_sales['date'].tail(14), daily_sales['sales'].tail(14), label='Actual Sales (Last 14 Days)', marker='o')
plt.plot(forecast_df['date'], forecast_df['predicted_sales'], label='Predicted Sales (Next 7 Days)', marker='s', linestyle='--', color='red')
plt.title('7-Day Sales Forecast')
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
import numpy as np

# 1. แบ่งข้อมูลเป็นส่วนเรียนรู้ 80% และส่วนทดสอบความแม่นยำ 20%
train_size = int(len(data_prepared) * 0.8)
train, test = data_prepared.iloc[:train_size], data_prepared.iloc[train_size:]

# 2. ฝึกโมเดลด้วยข้อมูล Train
X_train, y_train = train[features_list], train['sales']
X_test, y_test = test[features_list], test['sales']

model_eval = RandomForestRegressor(n_estimators=200, random_state=42)
model_eval.fit(X_train, y_train)

# 3. ให้โมเดลลองทายข้อมูลส่วน Test
y_pred = model_eval.predict(X_test)

# 4. คำนวณค่าความผิดพลาด
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)

print(f"--- รายงานความแม่นยำของโมเดล ---")
print(f"1. ความผิดพลาดเฉลี่ยต่อวัน (MAE): {mae:.2f} บาท")
print(f"2. ความผิดพลาดเป็นเปอร์เซ็นต์ (MAPE): {mape*100:.2f}%")
print(f"3. ความแม่นยำโดยประมาณ (Accuracy): {100 - (mape*100):.2f}%")

# ดูว่าวันไหนทายพลาดเยอะสุด
error_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred, 'Date': test['date']})
error_df['Error'] = error_df['Actual'] - error_df['Predicted']
print("\n--- ตัวอย่างวันที่ทายพลาด (Diff) ---")
print(error_df.head())
```

```python
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_percentage_error

# เตรียมลิสต์ของโมเดลที่จะใช้แข่งกัน
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(n_estimators=200, random_state=42),
    "XGBoost": XGBRegressor(n_estimators=200, learning_rate=0.05, random_state=42)
}

results = []

print(f"{'Model Name':<20} | {'Accuracy (%)':<15} | {'Error (Baht)':<15}")
print("-" * 55)

for name, model in models.items():
    # 1. สอนโมเดล
    model.fit(X_train, y_train)

    # 2. ให้โมเดลทาย
    y_pred = model.predict(X_test)

    # 3. วัดผล
    mape = mean_absolute_percentage_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    accuracy = 100 - (mape * 100)

    results.append({"Model": name, "Accuracy": accuracy, "MAE": mae})
    print(f"{name:<20} | {accuracy:<15.2f} | {mae:<15.2f}")

# หาตัวที่แม่นที่สุด
best_model = max(results, key=lambda x: x['Accuracy'])
print(f"\n สรุป: โมเดลที่แม่นที่สุดสำหรับร้านพี่คือ 👉 {best_model['Model']} 👈")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from datetime import timedelta
import matplotlib.pyplot as plt

# 1. Load data
df = pd.read_excel('/content/Coffee Shop Sales.xlsx')
df['total_sales'] = df['transaction_qty'] * df['unit_price']
df['transaction_date'] = pd.to_datetime(df['transaction_date'])

# 2. Daily Sales
daily_sales = df.groupby('transaction_date')['total_sales'].sum().reset_index()
daily_sales.columns = ['date', 'sales']
daily_sales = daily_sales.sort_values('date')
```

```python
# 3. Feature Engineering for Training
def create_features(df_in):
    df_out = df_in.copy()
    df_out['day_of_week'] = df_out['date'].dt.dayofweek
    df_out['month'] = df_out['date'].dt.month
    df_out['is_weekend'] = df_out['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
    df_out['sales_lag1'] = df_out['sales'].shift(1)
    df_out['sales_lag7'] = df_out['sales'].shift(7)
    df_out['rolling_mean_7'] = df_out['sales'].shift(1).rolling(window=7).mean()
    return df_out.dropna()

data_prepared = create_features(daily_sales)
features_list = ['day_of_week', 'month', 'is_weekend', 'sales_lag1', 'sales_lag7', 'rolling_mean_7']
```

```python
# 4. Train Winner Model (Linear Regression)
model = LinearRegression()
model.fit(data_prepared[features_list], data_prepared['sales'])
```

```python
# 5. 7-Day Forecasting (Recursive)
forecast_results = []
# Start from the last 7 days of actual sales to predict day 1
current_history = list(daily_sales['sales'].tail(7))
last_date = daily_sales['date'].max()

for i in range(1, 8):
    next_date = last_date + timedelta(days=i)
    # Prepare features for the next day
    next_features = pd.DataFrame([{
        'day_of_week': next_date.dayofweek,
        'month': next_date.month,
        'is_weekend': 1 if next_date.dayofweek >= 5 else 0,
        'sales_lag1': current_history[-1],
        'sales_lag7': current_history[0],
        'rolling_mean_7': np.mean(current_history)
    }])

    pred_sales = model.predict(next_features[features_list])[0]
    forecast_results.append({'date': next_date, 'predicted_sales': pred_sales})

    # Update history
    current_history.append(pred_sales)
    current_history.pop(0)

forecast_df = pd.DataFrame(forecast_results)
```

```python
# 6. Plot
plt.figure(figsize=(12, 6))
# Show last 30 days for context
plot_history = daily_sales.tail(30)
plt.plot(plot_history['date'], plot_history['sales'], label='Actual Sales (Last 30 Days)', marker='o', color='blue')
plt.plot(forecast_df['date'], forecast_df['predicted_sales'], label='7-Day Forecast', marker='s', linestyle='--', color='red')
plt.title('Coffee Shop 7-Day Sales Forecast (using Linear Regression)')
plt.xlabel('Date')
plt.ylabel('Sales Amount ($)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('sales_forecast_7days.png')
```

```python
import streamlit as st
import pandas as pd
import joblib
import numpy as np
from datetime import timedelta
import plotly.express as px

# ตั้งค่าหน้าเว็บ
st.set_page_config(page_title="Cafe Sales AI Forecast", layout="wide")

# 1. Load Model & Data
@st.cache_resource
def load_assets():
    model = joblib.load('coffee_sales_model.pkl')
    features = joblib.load('model_features.pkl')
    # โหลดข้อมูลมาทำ Insight สินค้า
    df = pd.read_excel('Coffee Shop Sales.xlsx')
    df['total_sales'] = df['transaction_qty'] * df['unit_price']
    df['transaction_date'] = pd.to_datetime(df['transaction_date'])
    return model, features, df

model, features_list, df = load_assets()

# --- ส่วนหัวข้อ ---
st.title("☕ Cafe Sales Forecast AI Dashboard")
st.markdown("ระบบพยากรณ์ยอดขายและวางแผนสต็อกวัตถุดิบอัตโนมัติ")

# --- ส่วนของการคำนวณ ---
daily_sales = df.groupby('transaction_date')['total_sales'].sum().reset_index()
last_date = daily_sales['transaction_date'].max()
history = list(daily_sales['total_sales'].tail(7))

# พยากรณ์ 7 วัน
forecast_list = []
curr_history = history.copy()
for i in range(1, 8):
    f_date = last_date + timedelta(days=i)
    input_data = pd.DataFrame([{
        'day_of_week': f_date.dayofweek,
        'month': f_date.month,
        'is_weekend': 1 if f_date.dayofweek >= 5 else 0,
        'sales_lag1': curr_history[-1],
        'sales_lag7': curr_history[0],
        'rolling_mean_7': np.mean(curr_history)
    }])
    pred = model.predict(input_data[features_list])[0]
    forecast_list.append({'Date': f_date, 'Forecast': pred})
    curr_history.append(pred)
    curr_history.pop(0)

forecast_df = pd.DataFrame(forecast_list)

# --- ส่วนการแสดงผล (UI) ---
col1, col2 = st.columns([2, 1])

with col1:
    st.subheader("📈 พยากรณ์ยอดขาย 7 วันข้างหน้า")
    fig = px.line(forecast_df, x='Date', y='Forecast', markers=True,
                  title="Sales Trend (Next 7 Days)",
                  labels={'Forecast': 'ยอดขายคาดการณ์ ($)'})
    st.plotly_chart(fig, use_container_width=True)

with col2:
    st.subheader("📦 สินค้าที่ต้องสต็อก (พรุ่งนี้)")
    next_day_name = (last_date + timedelta(days=1)).day_name()
    df['day_name'] = df['transaction_date'].dt.day_name()
    top_items = df[df['day_name'] == next_day_name].groupby('product_detail')['transaction_qty'].sum().nlargest(5)

    for item, qty in top_items.items():
        st.info(f"**{item}**: ประมาณ {int(qty/24)} - {int(qty/20)} หน่วย") # ปรับสูตรตามจริง

# --- ตารางสรุป ---
st.subheader("📋 รายละเอียดพยากรณ์รายวัน")
st.dataframe(forecast_df.style.format({'Forecast': '{:,.2f}'}), use_container_width=True)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import joblib

# โหลดข้อมูลเตรียมเทรนแบบเร็วๆ
df = pd.read_excel('Coffee Shop Sales.xlsx')
df['total_sales'] = df['transaction_qty'] * df['unit_price']
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
daily = df.groupby('transaction_date')['total_sales'].sum().reset_index()

# สร้าง Feature พื้นฐาน
daily['day_of_week'] = daily['transaction_date'].dt.dayofweek
daily['month'] = daily['transaction_date'].dt.month
daily['is_weekend'] = daily['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
daily['sales_lag1'] = daily['total_sales'].shift(1)
daily['sales_lag7'] = daily['total_sales'].shift(7)
daily['rolling_mean_7'] = daily['total_sales'].shift(1).rolling(window=7).mean()
daily = daily.dropna()

features_list = ['day_of_week', 'month', 'is_weekend', 'sales_lag1', 'sales_lag7', 'rolling_mean_7']
model = LinearRegression()
model.fit(daily[features_list], daily['total_sales'])

# บันทึกไฟล์สมอง AI
joblib.dump(model, 'coffee_model.pkl')
joblib.dump(features_list, 'features.pkl')
print("✅ บันทึกโมเดลสำเร็จ!")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
import numpy as np
from datetime import timedelta
import plotly.express as px
import os

# ตั้งค่าหน้าเว็บให้ดูทันสมัย
st.set_page_config(page_title="Cafe AI POS System", layout="wide", initial_sidebar_state="expanded")

# --- ฟังก์ชันหลัก: โหลดข้อมูล ---
@st.cache_resource
def load_data():
    file_name = 'Coffee Shop Sales.xlsx'
    if not os.path.exists(file_name):
        st.error(f"❌ ไม่พบไฟล์ {file_name}")
        st.stop()
    df = pd.read_excel(file_name)
    df['total_sales'] = df['transaction_qty'] * df['unit_price']
    df['transaction_date'] = pd.to_datetime(df['transaction_date'])
    return df

df = load_data()
model = joblib.load('coffee_model.pkl')
features_list = joblib.load('features.pkl')

# --- ส่วนของเมนู (SIDEBAR) ตามรูปพี่ ---
with st.sidebar:
    st.markdown("<h1 style='text-align: center; color: #FF4B4B;'>☕ CAFE AI</h1>", unsafe_allow_html=True)
    st.markdown("---")
    # เมนูที่พี่ระบุมา: แดชบอร์ด, บันทึกข้อมูล, ประวัติยอดขาย, คาดการณ์ยอดขาย
    menu = st.radio("เลือกรายการเมนู",
                    ["📊 แดชบอร์ด", "📝 บันทึกข้อมูล", "📜 ประวัติยอดขาย", "🔮 คาดการณ์ยอดขาย"],
                    index=0)
    st.markdown("---")
    if st.button("🚪 ออกจากระบบ", use_container_width=True):
        st.toast("ออกจากระบบเรียบร้อยแล้ว!")

# --- หน้าที่ 1: แดชบอร์ด (สรุปตัวเลข 3 ช่องตามรูป) ---
if menu == "📊 แดชบอร์ด":
    st.title("📊 สรุปยอดขายร้านวันนี้")

    last_date = df['transaction_date'].max()
    today_data = df[df['transaction_date'] == last_date]
    today_sales = today_data['total_sales'].sum()
    total_30d = df[df['transaction_date'] > (last_date - timedelta(days=30))]['total_sales'].sum()
    avg_daily = total_30d / 30

    # แสดงผล Metric Cards
    m1, m2, m3 = st.columns(3)
    m1.metric("ยอดขายวันนี้ (บาท)", f"฿{today_sales:,.2f}", f"{(today_sales-avg_daily):,.0f} จากค่าเฉลี่ย")
    m2.metric("ยอดรวม 30 วันล่าสุด", f"฿{total_30d:,.2f}")
    m3.metric("ยอดขายเฉลี่ยรายวัน", f"฿{avg_daily:,.2f}")

    st.markdown("---")

    # รายการของวันนี้ (เหมือนในรูปหน้าจอที่พี่ส่งมา)
    st.subheader("📋 รายการขายล่าสุดของวันนี้")
    st.dataframe(today_data[['transaction_time', 'product_detail', 'transaction_qty', 'total_sales']]
                 .sort_values('transaction_time', ascending=False), use_container_width=True)

# --- หน้าที่ 2: บันทึกข้อมูล ---
elif menu == "📝 บันทึกข้อมูล":
    st.title("📝 บันทึกข้อมูลการขายใหม่")
    with st.container(border=True):
        col_in1, col_in2 = st.columns(2)
        with col_in1:
            prod = st.selectbox("เลือกสินค้า", df['product_detail'].unique())
            qty = st.number_input("จำนวนที่ขายได้", min_value=1, step=1)
        with col_in2:
            price = st.number_input("ราคาต่อหน่วย", value=float(df[df['product_detail']==prod]['unit_price'].iloc[0]))
            st.write(f"ยอดรวม: **฿{qty * price:,.2f}**")

        if st.button("✅ บันทึกยอดขาย", type="primary"):
            st.success(f"บันทึกข้อมูล {prod} จำนวน {qty} รายการ สำเร็จ!")

# --- หน้าที่ 3: ประวัติยอดขาย ---
elif menu == "📜 ประวัติยอดขาย":
    st.title("📜 ประวัติยอดขายย้อนหลัง")
    # ตัวกรองวันที่
    date_range = st.date_input("เลือกช่วงเวลา", [df['transaction_date'].min(), df['transaction_date'].max()])
    filtered_df = df[(df['transaction_date'] >= pd.to_datetime(date_range[0])) &
                     (df['transaction_date'] <= pd.to_datetime(date_range[1]))]
    st.write(f"พบข้อมูลทั้งหมด {len(filtered_df)} รายการ")
    st.dataframe(filtered_df.sort_values('transaction_date', ascending=False), use_container_width=True)

# --- หน้าที่ 4: คาดการณ์ยอดขาย (พลังจาก AI 93.54%) ---
elif menu == "🔮 คาดการณ์ยอดขาย":
    st.title("🔮 ระบบพยากรณ์ยอดขาย (AI Forecast)")
    st.info("🤖 โมเดลใช้ข้อมูลประวัติย้อนหลัง 7 วันล่าสุด เพื่อประมวลผลด้วยความแม่นยำ 93.54%")

    daily_sales = df.groupby('transaction_date')['total_sales'].sum().reset_index()
    history = list(daily_sales['total_sales'].tail(7))
    forecast = []

    # วนลูปพยากรณ์ 7 วัน
    for i in range(1, 8):
        f_date = daily_sales['transaction_date'].max() + timedelta(days=i)
        inp = pd.DataFrame([{
            'day_of_week': f_date.dayofweek, 'month': f_date.month,
            'is_weekend': 1 if f_date.dayofweek >= 5 else 0,
            'sales_lag1': history[-1], 'sales_lag7': history[0],
            'rolling_mean_7': np.mean(history)
        }])
        pred = model.predict(inp[features_list])[0]
        forecast.append({'วันที่': f_date.date(), 'ยอดคาดการณ์ (บาท)': round(pred, 2)})
        history.append(pred)
        history.pop(0)

    # แสดงกราฟและตาราง
    f_df = pd.DataFrame(forecast)
    st.plotly_chart(px.line(f_df, x='วันที่', y='ยอดคาดการณ์ (บาท)', markers=True, title="แนวโน้มยอดขาย 7 วันข้างหน้า"))
    st.table(f_df)
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
import numpy as np
from datetime import timedelta
import plotly.express as px
import os

# --- 1. ตั้งค่าความปลอดภัย (Login) ---
def login():
    st.markdown("<h2 style='text-align: center;'>🔐 Cafe AI Login</h2>", unsafe_allow_html=True)
    user = st.text_input("Username")
    pw = st.text_input("Password", type="password")
    if st.button("Login"):
        if user == "admin" and pw == "1234": # พี่เปลี่ยน User/Pass ตรงนี้ได้ครับ
            st.session_state['logged_in'] = True
            st.rerun()
        else:
            st.error("ชื่อผู้ใช้หรือรหัสผ่านไม่ถูกต้อง")

if 'logged_in' not in st.session_state:
    st.session_state['logged_in'] = False

if not st.session_state['logged_in']:
    login()
    st.stop()

# --- 2. เริ่มส่วนของแอปหลัก ---
st.set_page_config(page_title="Cafe Management App", layout="wide")

@st.cache_resource
def load_assets():
    df = pd.read_excel('Coffee Shop Sales.xlsx')
    df['total_sales'] = df['transaction_qty'] * df['unit_price']
    df['transaction_date'] = pd.to_datetime(df['transaction_date'])
    model = joblib.load('coffee_model.pkl')
    features = joblib.load('features.pkl')
    return df, model, features

df, model, features_list = load_assets()

# --- 3. SIDEBAR เมนูตามรูปพี่ ---
with st.sidebar:
    st.title("☕ CAFE SYSTEM")
    menu = st.radio("เมนูหลัก", ["📊 แดชบอร์ดสรุปผล", "🔮 คาดการณ์ยอดขาย", "📝 บันทึกข้อมูล", "📜 ประวัติยอดขาย"])
    st.divider()
    if st.button("🚪 ออกจากระบบ"):
        st.session_state['logged_in'] = False
        st.rerun()

# --- 4. หน้าแดชบอร์ด ---
if menu == "📊 แดชบอร์ดสรุปผล":
    st.header("📊 สรุปยอดขายร้านวันนี้")
    last_date = df['transaction_date'].max()
    today_sales = df[df['transaction_date'] == last_date]['total_sales'].sum()
    total_30d = df[df['transaction_date'] > (last_date - timedelta(days=30))]['total_sales'].sum()
    avg_daily = total_30d / 30

    c1, c2, c3 = st.columns(3)
    c1.metric("ยอดขายวันนี้", f"฿{today_sales:,.2f}")
    c2.metric("ยอดขายรวม 30 วัน", f"฿{total_30d:,.2f}")
    c3.metric("ยอดขายเฉลี่ยรายวัน", f"฿{avg_daily:,.2f}")

    st.subheader("📋 รายการขายล่าสุดของวันนี้")
    st.dataframe(df[df['transaction_date'] == last_date].sort_values('transaction_time', ascending=False), use_container_width=True)

# --- 5. หน้าคาดการณ์ยอดขาย ---
elif menu == "🔮 คาดการณ์ยอดขาย":
    st.header("🔮 AI Forecast (Accuracy 93.54%)")
    daily_sales = df.groupby('transaction_date')['total_sales'].sum().reset_index()
    history = list(daily_sales['total_sales'].tail(7))
    forecast = []

    for i in range(1, 8):
        f_date = daily_sales['transaction_date'].max() + timedelta(days=i)
        inp = pd.DataFrame([{
            'day_of_week': f_date.dayofweek, 'month': f_date.month,
            'is_weekend': 1 if f_date.dayofweek >= 5 else 0,
            'sales_lag1': history[-1], 'sales_lag7': history[0],
            'rolling_mean_7': np.mean(history)
        }])
        pred = model.predict(inp[features_list])[0]
        forecast.append({'วันที่': f_date.date(), 'พยากรณ์ยอดขาย': round(pred, 2)})
        history.append(pred)
        history.pop(0)

    st.plotly_chart(px.line(pd.DataFrame(forecast), x='วันที่', y='พยากรณ์ยอดขาย', markers=True))
    st.table(pd.DataFrame(forecast))
```
